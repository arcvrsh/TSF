```python
# import all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('ggplot') # theme
```

```python
cric_data = pd.read_csv("dataset/matches.csv")
cric_data
```

```python
# Drop uneccessary column
cols_to_drop = ['dl_applied']
cric_data.drop(cols_to_drop,axis=1,inplace=True)
cric_data
```

# No. of Matches played in each Season

```python
plt.figure(figsize=(15,5))
sns.set_theme(style='dark')
sns.countplot(x='season',saturation=0.95,data=cric_data)
plt.xticks(fontname='Times new Roman',fontsize=15)
plt.yticks(fontsize=15,fontname='Times New Roman')
plt.xlabel('Season',fontname='Times New Roman',fontsize=22)
plt.ylabel('Count',fontname='Times New Roman',fontsize=22)
plt.title('Matches Played in Each Season ',fontname='Times New Roman',fontsize=42)
plt.show()
```

# Win Percentage Field First

```python
total_wins = (cric_data.win_by_wickets>0).sum()
total_loss = (cric_data.win_by_wickets==0).sum()
total = total_wins + total_loss
percent = [(total_wins/total)*100,(total_loss/total)*100]
plt.figure(figsize=(20,8))
plt.pie(percent,
        labels=('Wins','Loss'),
        autopct='%.1f%%',
        colors=('DarkOliveGreen','Yellow'),
        startangle=90,
        shadow=True,
        textprops={'fontsize':30,'fontname':'Times New Roman'})

plt.axis('equal')
plt.title("Win Percentage Field First",fontname='Times New Roman',fontsize=42)
plt.show()
```

## Almost 53.3 times teams choose field first and won the match

```python
del_df=pd.read_csv("dataset/deliveries.csv")
del_df
```

```python
del_df.info()
```

```python
del_df.isnull().sum()
```

```python
cols_drop=['player_dismissed','dismissal_kind','fielder']
del_df.drop(cols_drop,axis=1,inplace=True)
del_df.head()
```

```python
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.02*height,
                '%d' % int(height),
                ha='center', va='bottom')
```

```python
temp_df = del_df.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
temp_df = temp_df.iloc[:10,:]

labels = np.array(temp_df['batsman'])
ind = np.arange(len(labels))
fig, ax = plt.subplots(figsize=(15,6))
width=0.5
rects = ax.bar(ind, np.array(temp_df['batsman_runs']), width=width, color='cornflowerblue')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical',fontsize=22,fontname="Times New Roman")
ax.set_ylabel("Count",fontsize=22,fontname="Times New Roman")
ax.set_title("Top run scorers in IPL",fontsize=42,fontname="Times New Roman")
autolabel(rects)
ax.set_facecolor("gray")
plt.show()
```

## Suresh Raina has most runs in IPL that's why he is known from name MR.IPL followed by Virat Kohli who is behind just few runs from him

# Batsman with most no. of Runs

```python
temp_df = del_df.groupby('batsman')['batsman_runs'].agg(lambda x: (x==4).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
temp_df = temp_df.iloc[:10,:]

labels = np.array(temp_df['batsman'])
ind = np.arange(len(labels))
fig, ax = plt.subplots(figsize=(15,6))
width=0.5
rects = ax.bar(ind, np.array(temp_df['batsman_runs']), width=width, color='olive')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical',fontsize=22,fontname="Times New Roman")
ax.set_ylabel("Count",fontsize=22,fontname="Times New Roman")
ax.set_title("Batsman with most no. of boundaries",fontsize=42,fontname="Times New Roman")
autolabel(rects)
ax.set_facecolor("gray")
plt.show()
```

## Gautam Gambhir score most boundaries followed by Shikhar Dhawan
